```python
def get_batched_timing(algo, task):
    if algo!="serial":
        batched_timings = np.stack([timings[task][algo][str(i)] for i in range(1,7)]).T
    else:
        batched_timings = np.stack([timings[task][algo] for i in range(1,7)]).T
    return batched_timings
```

```python
def get_sq_batched_timing(algo, task):
    if algo!="serial":
        batched_timings = np.stack([timings[task][algo][str(i)] for i in [1,4,9]]).T
    else:
        batched_timings = np.stack([timings[task][algo] for i in [1,4,9]]).T
    return batched_timings
```

```python
def batch_get_mean_std(data):
    return np.mean(data, axis = 0), np.std(data, axis = 0)
```

Timing

```python
fig, ax = plt.subplots()
ax.set_xlabel("threads")
ax.set_title("Execution time for 100000x1000 matrix")
for alg in ["serial", "rows", "cols"]:
    m, v = batch_get_mean_std(get_batched_timing(alg, "100000 1000"))
    ax.errorbar(range(1,7), m, yerr=v, label=alg)
fig.legend()
plt.show()
```
